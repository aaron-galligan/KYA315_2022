```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

## Velocity potential (2D)

The velocity potential for a point source and a point sink equidistant from origin is given as

$$ \Phi(\mathbf{r}) = - \frac{m}{4\pi\sqrt{(x+d)^2+y^2}} + \frac{m}{4\pi\sqrt{(x-d)^2+y^2}} $$

Let $m = 1/d$:

```python
def potential(x,y,d): 
    Phi = -1/(d*4*np.pi*((x+d)**2+y**2)**(1/2)) + 1/(d*4*np.pi*((x-d)**2+y**2)**(1/2))
    return Phi
```

The velocity vector is then $\mathbf{u}=\nabla\Phi$ or

$$ \mathbf{u} = \begin{bmatrix} \frac{(x+d)m}{4\pi((x+d)^2+y^2)^{\frac{3}{2}}} - \frac{(x-d)m}{4\pi((x-d)^2+y^2)^{\frac{3}{2}}} \\ 
\frac{ym}{4\pi((x+d)^2+y^2)^{\frac{3}{2}}} - \frac{ym}{4\pi((x-d)^2+y^2)^{\frac{3}{2}}} \end{bmatrix} $$

```python
def velocity(x,y,d):
    u = (x+d)/(d*4*np.pi*((x+d)**2+y**2)**(3/2)) - (x-d)/(d*4*np.pi*((x-d)**2+y**2)**(3/2))
    v = y/(d*4*np.pi*((x+d)**2+y**2)**(3/2)) - y/(d*4*np.pi*((x-d)**2+y**2)**(3/2))
    U = np.array([u,v])
    return U
```

```python
velocity(0,0,1)
```

## Let $d=1$

```python
d = 1
```

### Calculate velocity on gridded domain

```python
x = np.linspace(-1.5,1.5,20)
y = np.linspace(-1.5,1.5,20)
X,Y = np.meshgrid(x,y)
```

```python
U = velocity(X,Y,d)
```

```python
u = U[0,:,:]
v = U[1,:,:]
```

### Calculate velocity potential

```python
Phi = potential(X,Y,d)
```

### Plot velocity potential and streamlines

```python
fig = plt.figure(figsize=(10,10))
p = plt.contourf(X,Y,Phi,cmap='RdBu',levels=100)
plt.streamplot(X,Y,u,v,color='k');
plt.colorbar(p)
```

## Velocity potential (2D) - limiting case

The velocity potential for a point source and a point sink at the origin is given as

$$ \Phi(\mathbf{r}) = \frac{\mu x}{4\pi(x^2+y^2)^{3/2}} $$

```python
def potential(x,y,mu): 
    Phi = mu*x/(4*np.pi*(x**2+y**2)**(3/2))
    return Phi
```

The velocity vector is then $\mathbf{u}=\nabla\Phi$ or

$$ \mathbf{u} = \begin{bmatrix} \frac{\mu}{4\pi(x^2+y^2)^{3/2}} - \frac{3\mu x^2}{4\pi(x^2+y^2)^{5/2}} \\ 
-\frac{3\mu xy}{4\pi(x^2+y^2)^{5/2}} \end{bmatrix} $$

```python
def velocity(x,y,mu):
    u = mu/(4*np.pi*(x**2+y**2)**(3/2)) - 3*mu*x**2/(4*np.pi*(x**2+y**2)**(5/2))
    v = -3*mu*x*y/(4*np.pi*(x**2+y**2)**(5/2))
    U = np.array([u,v])
    return U
```

```python
velocity(0,1,1)
```
